```python
#import libraries
import matplotlib.pyplot as plt
```

```python
#set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul","Aug","Sep", "Oct", "Nov", "Dec"]
 
#Set the y_axis to a list of floats as the total fare int he US dollars accumulated for each month.
y_axis = [10.02,23.24,39.20,35.42,32.34,27.04,43.82,10.56,11.85,27.90,20.71,20.09]
```

```python
#create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
#create the plot with ax.plt()
#fig = plt.figure()
#ax= fig.add_subplot()

#fig, ax = plt.subplots()

ax = plt.axes()

ax.plot(x_axis, y_axis)
```

```python
#create the plot and add a label for the legend
plt.plot(x_axis, y_axis,marker ='*', color='blue', linewidth=2, label='Boston')
#create labels for the axis
plt.xlabel("Date")
plt.ylabel("Fare ($)")
#set the y limit between 0 and 45
plt.ylim(0,45)
#create a title
plt.title("Pyber Fare by Month")
plt.grid()
#Add a legend
plt.legend()
```

```python
#create a bar plot
plt.barh( x_axis, y_axis,  color='magenta', label='boston')
plt.gca().invert_yaxis()
#create labels for axia
plt.xlabel('Fare (S)')
plt.ylabel('Date')
#create title
plt.title('PyBer Fare by Month')
#add a legend
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()

ax.barh(x_axis ,y_axis, color="cyan", label = "Chicago")

ax.legend()
```

```python
#create a scatter plot 
plt.scatter( y_axis, x_axis, c="purple", marker = "o", label="Chicago")
plt.xlabel("Fare ($)")
plt.ylabel("Month")
plt.gca().invert_yaxis()
plt.title("PyBer Fare by Month")
plt.xlim(0,50)
plt.legend()
```

```python
#create a bubble chart
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)

```

```python
#create a bubble chart with size changed using list comprehension
#new_list = [expression for item in list if conditional]
plt.scatter(x_axis, y_axis, s=[i*3 for i in y_axis])
```

```python
#create scatter plot using Object Oriented Method
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
#create a bubble chart using object oriented method
fig, ax = plt.subplots()
ax.scatter(y_axis,x_axis, s=[i*5 for i in y_axis], c='skyblue', marker='o', alpha = .8, linewidths=2, edgecolor = "black", label = 'Boston' )
plt.gca().invert_yaxis()
ax.legend()
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Date')
ax.set_xlim(0,50)
```

```python
#create pie charg using matlab method
plt.subplots(figsize=(8,8))
colors = ['slateblue','magenta','lightblue','green','yellowgreen','greenyellow','yellow','orange','gold','indianred','tomato','mistyrose']
explode_values = (0,0,0.1,0,0,0,0.2,0,0,0,0,0)
plt.pie(y_axis, explode = explode_values, colors = colors, labels = x_axis, autopct = '%.1f%%')
plt.show
```

```python
#create a pie chart using object oriented method
fig, ax=plt.subplots(figsize=(8,8))

ax.pie(y_axis, labels=x_axis, explode = explode_values, autopct = '%.1f%%', startangle = 90, counterclock = False, shadow=True, colors = colors)
plt.show()
```
